## Content based recommender system 

Let us try to implement some of the strategies seen during the lectures to provide more personalized recommendations. We start from a content based approach. We will start implementing the recommnender for a specific user. We will see the following steps:

- Selection of a random user
- Retrieving the information of the rated and unrated movies for the selected user
- Compute a feature representation for the movie's contents (TF-IDF)
- Train a regressor, computing of the predicted ratings for unrated movies, and recommending the movies with higher predicted ratings


1. Selection of a random user
For simplicity, we select a random user from the filtered_df Dataframe having more than 200 evaluations. 
First, we select all the users with more than 200 evaluations, and then we use the sample method to extract a random row. Finally, we extract the user column and we read the content with the iloc method.

```python
# user_df = pd.read_csv("/Users/thaole26601/Downloads/RAW_interactions.csv")
# user_df.head()
# Header = ['user', 'item', 'date', 'rating', 'review']
# user_df.columns = Header


users_ratings = filtered_df.groupby(['user']).count()
selected = users_ratings['rating'] > 200
selected_users = users_ratings.loc[selected]
random_selected = selected_users.sample() # sample() returns a random row from the dataframe. The returned object is a dataframe with one row. If you pass a number as argument you specify to select more than one row.
select_column_df = random_selected.reset_index()['user'] # reset_index() create a new index, and the user became a column. Then, we can filter using the column name
selected_user = select_column_df.iloc[0] # iloc select by index, since our dataframe only has one row we read it from the index 0
print("Selected user: " + str(selected_user))
```

```python
#2. Retrieving the information of the rated and unrated recipes for the selected user.

selected_user_ratings = filtered_df.loc[filtered_df['user'] == selected_user]
selected_user_ratings = selected_user_ratings.sort_values(by='item', ascending=True)
print("Rated recipes: " + str(selected_user_ratings.shape[0]))
display(selected_user_ratings.head(10))
```

Now, we need the select content of the rated recipes, that we will need to compute the features that our model will use. 
- Select the rows from our recipe_df dataframe having the item that is in the list of item of the selected_user_ratings Dataframe, which only contains the evaluations of the considered user. 
- We do this using the isin method. Since the item is the index in recipes_df, we can select directly pass the list of item we want to select to loc, and it will return the rows for which the index is in the considered list. Finally, we only select the columns name and description, ingredients.

```python
rated_recipe_df = recipe_df.loc[recipe_df['item'].isin(list(selected_user_ratings['item']))]

rated_recipe_df = rated_recipe_df.merge(selected_user_ratings, on='item')
rated_recipe_df['rating']= rated_recipe_df['rating'].transform(lambda x: x - x.mean())
print("Rated recipes: " + str(rated_recipe_df.shape[0]))
display(rated_recipe_df.head(10))
```

```python
diff = set(recipe_df.index) - set(rated_recipe_df.index)
unrated_recipe_df = recipe_df.loc[diff]

unrated_recipe_df = unrated_recipe_df.sample(n=500)
print("Unrated recipes: " + str(unrated_recipe_df.shape[0]))
unrated_recipe_df = unrated_recipe_df.merge(filtered_df, on='item')
unrated_recipe_df['rating']= unrated_recipe_df['rating'].transform(lambda x: x - x.mean())
display(unrated_recipe_df.head(3))

```

```python
X_test = unrated_recipe_df[features_columns] # x contains the columns of the selected features
y_test = unrated_recipe_df['rating'] # y contains the classification feature.
```

```python
y_unrated = neigh.predict(X_test)
```

```python
unrated_recipe_df['predicted_ratings_kNN'] = y_unrated
unrated_recipe_df_knn_ordering = unrated_recipe_df.sort_values(by='predicted_ratings_kNN', ascending=False)
display(unrated_recipe_df_knn_ordering.head(3))

mean_squared_error(unrated_recipe_df['rating'], unrated_recipe_df['predicted_ratings_kNN'])
```

### Evaluation of CF RS

We see now an example of evaluation of a collaborative filtering approach. We will use again the Lenskit library for this. Since we have one single model for all the users, the evaluation is simpler: we can just consider all the ratings we have and split them into training and test set.; then, we train the recommender on the trainig set, and we use it to predict the results on the test set. We will do it using the predict method. Finally, we compute the relevance values from the original ratings and from the predicted ratings, and we compare them to obtain precision, recall and fscore values.



```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support

train_df, test_df = train_test_split(filtered_df, test_size=0.2)

item_item = ItemItem(10, min_nbrs=1)  # Minimum (3) and maximum (15) number of neighbors to consider
recsys = Recommender.adapt(item_item)
recsys.fit(train_df)

test_df['predicted_rating'] = recsys.predict(test_df)

test_df['relevant'] = test_df['rating'].apply(lambda x: 1 if x>3 else 0)
test_df['predicted_relevant'] = test_df['predicted_rating'].apply(lambda x: 1 if x>3 else 0)

y_test = list(test_df['relevant'])
y_pred = list(test_df['predicted_relevant'])
precision, recall, fscore, _ = precision_recall_fscore_support(y_test, y_pred, average="binary")

print("Precision:\t" + str(precision) + 
      "\nRecall:\t" + str(recall) + 
      "\nFscore:\t" + str(fscore))

#precision and recall increase after min_nbrs is reduced
```

### Group recommender system 

goals: Implement the basic agregation strategies for group recommendations. Generate a simple textual explanation for such strategies.

Summary

- Explanations for Group Recommenders




```python
#select a random set of five users from our dataset. users having at least 200 evaluations.
users_ratings = filtered_df.groupby(['user']).count()
selected = users_ratings['rating'] > 200
selected_users = users_ratings.loc[selected]
random_selected = selected_users.sample(n=5) # sample() returns now n random rows from the dataframe. The returned object is a dataframe with five rows. 
select_column_df = random_selected.reset_index()['user'] # reset_index() create a new index, and the userId became a column. Then, we can filter using the column name
group_users = list(select_column_df) 0
print(group_users)
print(selected_users)
print(selected)
```

```python
#recommend a list of 10 recipes that nobody has seen yet. Consider movies with at least 10 evaluations.
group_ratings = filtered_df.loc[filtered_df['user'].isin(group_users)]
all_recipes = set(recipe_df.index.tolist())
num_ratings_df = filtered_df.groupby(['item']).count()
considered_recipes = set(num_ratings_df.loc[num_ratings_df['user'] > 10].reset_index()['item'])

group_rated_recipes = set(group_ratings['item'].tolist())
group_unseen_recipes = considered_recipes - group_rated_recipes

print(len(all_recipes))
print(len(considered_recipes))
print(len(group_rated_recipes))
print(len(group_unseen_recipes))
```

```python
#Evaluate individuals' preferences for the unseen movies
import itertools

group_unseen_df = pd.DataFrame(list(itertools.product(group_users, group_unseen_recipes)), columns=['user', 'item'])
group_unseen_df['predicted_rating'] = recsys.predict(group_unseen_df)
group_unseen_df = group_unseen_df.loc[group_unseen_df['predicted_rating'].notnull()]
display(group_unseen_df)
```

```python
group_unseen_df.groupby('user').predicted_rating.mean()
```

### Aggregation Strategies for Group Recommenders
 

```python
#Additive strategy: the sum of all the individuals ratings

additive_df = group_unseen_df.groupby('item').sum()
additive_df = additive_df.join(recipe_df['name'], on='item')
additive_df = additive_df.sort_values(by="predicted_rating", ascending=False).reset_index()[['item', 'name', 'predicted_rating']]
display(additive_df)
```

```python
# least misery

least_misery_df = group_unseen_df.groupby('item').min()
least_misery_df = least_misery_df.join(recipe_df['name'], on='item')
least_misery_df = least_misery_df.sort_values(by="predicted_rating", ascending=False).reset_index()[['item', 'name', 'predicted_rating']]
display(least_misery_df.head(10))
```

```python
# most pleasure

most_pleasure_df = group_unseen_df.groupby('item').max()
most_pleasure_df = most_pleasure_df.join(recipe_df['name'], on='item')
most_pleasure_df = most_pleasure_df.sort_values(by="predicted_rating", ascending=False).reset_index()[['item', 'name', 'predicted_rating']]
display(most_pleasure_df.head(10))
```

```python
# Fairness

def generate_preference_list(user):
    individual_df = group_unseen_df.loc[group_unseen_df['user']==user]
    return list(individual_df.sort_values(by="predicted_rating", ascending=False).reset_index()['item'])

individual_preference_lists = dict()
for member in group_users:
    individual_preference_lists[member] = generate_preference_list(member)
    
result = list()
for i in range(10):
    user = group_users[i % 5]
    user_best = individual_preference_lists[user].pop(0)
    for member in group_users:
        if user_best in individual_preference_lists[member]:
            individual_preference_lists[member].remove(user_best)
    result.append(user_best)
    
fairness_df = pd.DataFrame(result, columns=['item']).join(recipe_df['name'], on='item')
display(fairness_df)
```

```python
def split_int(arr):
    split_recipes=[]
    #for i in range (len(arr)):

    to_split=arr
    print(type(to_split))
    test_split=(to_split.split(','))
        
    for i in range (len(test_split)):
        if i==0:
            
            split_recipes.append((int)(test_split[i][1:]))
            
        if i==len(test_split)-1:
            split_recipes.append((int)(test_split[len(test_split)-1][:len(test_split[len(test_split)-1])-1])) 

        elif i>0 & i<len(test_split)-2:
            split_recipes.append((test_split[i]))        

    return split_recipes
```

```python
def split_float(arr):
    split_ratings=[]
    #for i in range (len(arr)):

    to_split=arr
    print(type(to_split))
    test_split=(to_split.split(','))
        
    for i in range (len(test_split)):
        if i==0:
            
            split_ratings.append((float)(test_split[i][1:]))
            
        if i==len(test_split)-1:
            split_ratings.append((float)(test_split[len(test_split)-1][:len(test_split[len(test_split)-1])-1])) 

        elif i>0 & i<len(test_split)-2:
            split_ratings.append((test_split[i]))        

    return split_ratings
```

```python
# Content Based Filtering
class ContentBasedFiltering:
    def __init__(self, groupRecommendation):
        self.groupRecommendation=groupRecommendation

    # Obtains all the items interacted with and rating provided by user    
    def retrieve_info (self,user_id):
        selected_user_ratings = user_recipe_df.loc[(user_recipe_df['user']) == float(user_id)]
        #selected_user_ratings = selected_user_ratings.sort_values(by='item', ascending=True)
        return(selected_user_ratings)
    
    # Combines recipe rating value pairs for the entire group
    def combine_info( self,selected_users):
        group_contentbased_info=[]
        for user_id in selected_users:
            group_contentbased_info.append(self.retrieve_info(user_id))
        
        return (group_contentbased_info)

    # Retrieves recipe_id of the recipes interacted by user
    def recipe_info(self, user_rating):
        
        rated_recipe_df = df_vegan_recipe.loc[df_vegan_recipe['item'].isin(list(user_rating['item']))]
        rated_recipe_df['item']= rated_recipe_df['item'].astype('object')
        rated_recipe_df =  rated_recipe_df.merge(user_rating, on='item')
        rated_recipe_df['rating']= rated_recipe_df['rating'].transform(lambda x: x - x.mean())
        return (rated_recipe_df)
    
    # Combines all the recipes interacted by the group. Used to create a dataset with only unseen recipe
    def combine_recipe_info(self, selected_users):
        group_recipe_df=pd.DataFrame()
        for user_id in selected_users:
            user_rating= self.retrieve_info(user_id)
            user_recipe_info= self.recipe_info(user_rating)
            group_recipe_df= group_recipe_df.append(user_recipe_info)
        
        return (group_recipe_df)
    
    # Returs the recipe's that the group has not interacted with
    
    # The main method that combines all the information, trains a KNN regressor and to generate ratings and make recommendations
    def train_predict(self,selected_group):
        #group_rating= self.combine_info(selected_group)
        group_recipe_df= self.combine_recipe_info(selected_group)
        
        group_unrated_df= df_vegan_recipe


        features_columns = ['dessert'] # the features we have selected from the dataset
        X_train = group_recipe_df[features_columns] # x contains the columns of the selected features
        y_train = group_recipe_df['rating'] # y contains the classification feature.
        X_test = group_unrated_df[features_columns] # x contains the columns of the selected features
        #y_test = group_unrated_df['ratings'] # y contains the classification feature.


        neigh = KNeighborsRegressor(n_neighbors=3)

        
        neigh.fit(X_train, y_train) 
        y_unrated = neigh.predict(X_test)
        group_unrated_df['predicted_ratings_kNN'] = y_unrated
        groupRecommendation = group_unrated_df.sort_values(by='predicted_ratings_kNN', ascending=False)
        return (groupRecommendation)
           


            
            
```
